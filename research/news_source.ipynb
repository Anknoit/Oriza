```python
from bs4 import BeautifulSoup
import requests
# url = "https://www.spglobal.com/commodity-insights/en/news-research/latest-news#q=&rows=20&pagenum=1&sort=es_unified_dt%20desc"
url = "https://www.livemint.com/market/commodities"
response = requests.get(url)
html_content = response.text
html_content
soup = BeautifulSoup(html_content, "html.parser")
news = soup.find_all('h2')
for i in news:
    # print(i)
    news_text, news_link = i.get_text(), url+i.a['href']
    # print(news_text, news_link)
    news_list = {
        "news" : news_text,
        "link" : news_link,
        "source" : "mint"
    }
print(news_list)

    
```
